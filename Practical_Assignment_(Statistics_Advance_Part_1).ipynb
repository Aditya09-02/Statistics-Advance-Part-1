Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution ?

```python
#Ans.

# prompt: Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution ?

def bernoulli_pdf(p, k):
    """
    Calculates the probability distribution function (PDF) of a Bernoulli distribution.

    Args:
        p: The probability of success (a value between 0 and 1).
        k: The outcome (either 0 or 1).

    Returns:
        The probability of the given outcome k, or None if k is not 0 or 1.
    """
    if k == 0:
        return 1 - p
    elif k == 1:
        return p
    else:
        return None  # Invalid outcome
```

Q4. Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram ?

```python
#Ans.

# prompt: Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram ?

import random
import matplotlib.pyplot as plt

def binomial_simulation(n, p, num_trials):
    """Simulates a binomial distribution and returns the results."""
    results = []
    for _ in range(num_trials):
        successes = 0
        for _ in range(n):
            if random.random() < p:
                successes += 1
        results.append(successes)
    return results

# Parameters for the binomial distribution
n = 10  # Number of trials
p = 0.5  # Probability of success
num_trials = 1000  # Number of simulations

# Simulate the binomial distribution
outcomes = binomial_simulation(n, p, num_trials)


# Plot the histogram
plt.hist(outcomes, bins=range(n + 2), align='left', rwidth=0.8)  # Adjust bins for better visualization
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.xticks(range(n + 1))  # Ensure all possible outcomes are displayed on x-axis
plt.show()
```

Q6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution ?

```python
#Ans.

# prompt: Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution ?

import matplotlib.pyplot as plt
import numpy as np

def plot_cdf_discrete_uniform(low, high):
    """
    Calculates and plots the cumulative distribution function (CDF) of a discrete uniform distribution.

    Args:
      low: The lower bound of the uniform distribution (inclusive).
      high: The upper bound of the uniform distribution (inclusive).
    """

    # Generate the possible values and their probabilities
    values = np.arange(low, high + 1)
    probabilities = np.full(len(values), 1 / len(values))  # Probability for each value

    # Calculate the cumulative probabilities
    cumulative_probabilities = np.cumsum(probabilities)

    # Plot the CDF
    plt.figure(figsize=(8, 6))
    plt.step(values, cumulative_probabilities, where='post')
    plt.xlabel("Values")
    plt.ylabel("Cumulative Probability")
    plt.title(f"CDF of Discrete Uniform Distribution (low={low}, high={high})")
    plt.grid(True)
    plt.show()


# Example usage
plot_cdf_discrete_uniform(1, 6)  # Example: Uniform distribution from 1 to 6
```

Q10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution ?

```python
# Ans.

# prompt: Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution ?

import random
import matplotlib.pyplot as plt
import numpy as np

def clt_simulation(distribution, n_samples, n_repetitions):
    """
    Simulates the Central Limit Theorem for a given distribution.

    Args:
        distribution: A function that generates random numbers from the desired distribution.
        n_samples: The number of samples to draw from the distribution in each repetition.
        n_repetitions: The number of times to repeat the sampling process.

    Returns:
        A list of sample means.
    """
    sample_means = []
    for _ in range(n_repetitions):
        samples = [distribution() for _ in range(n_samples)]
        sample_means.append(np.mean(samples))
    return sample_means

# Example usage with an exponential distribution
def exponential_distribution(lambd=1):
  """Generates a random number from an exponential distribution."""
  return -np.log(1 - random.random()) / lambd

n_samples = 30  # Number of samples per repetition
n_repetitions = 1000  # Number of repetitions

# Simulate the CLT
sample_means = clt_simulation(exponential_distribution, n_samples, n_repetitions)

# Plot the histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue')
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.title(f"Distribution of Sample Means (n={n_samples})")
plt.grid(True)
plt.show()
```

Q14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution ?

```python
#Ans.

# prompt:  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution ?

def calculate_zscore(data_point, data_set):
    """
    Calculates the Z-score for a given data point compared to a dataset.

    Args:
      data_point: The data point for which to calculate the Z-score.
      data_set: A list or NumPy array of data points representing the dataset.

    Returns:
      The Z-score of the data point.
    """

    import numpy as np
    from scipy.stats import norm

    # Calculate the mean and standard deviation of the dataset
    mean = np.mean(data_set)
    std_dev = np.std(data_set)

    # Calculate the Z-score
    z_score = (data_point - mean) / std_dev

    # Compare to the standard normal distribution
    probability = norm.cdf(z_score)  # Cumulative distribution function

    print(f"Z-score: {z_score}")
    print(f"Probability (data point <= given value): {probability}")

    return z_score, probability

# Example Usage:
data = np.random.normal(loc=50, scale=10, size=100) # Example data set
data_point = 65

z_score, probability = calculate_zscore(data_point, data)
```

Q15.  Implement hypothesis testing using Z-statistics for a sample dataset ?

```python
#Ans.

# prompt: Implement hypothesis testing using Z-statistics for a sample dataset ?

from scipy.stats import norm

def z_test(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    """
    Performs a one-sample Z-test for a population mean.

    Args:
        sample_mean: The mean of the sample.
        population_mean: The hypothesized population mean.
        population_std: The population standard deviation.
        sample_size: The size of the sample.
        alpha: The significance level (default is 0.05).

    Returns:
        A dictionary containing the test statistic, p-value, and conclusion.
    """

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / (population_std / (sample_size**0.5))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    # Make a decision based on the p-value
    conclusion = "Fail to reject the null hypothesis"
    if p_value < alpha:
        conclusion = "Reject the null hypothesis"

    return {"z_statistic": z_statistic, "p_value": p_value, "conclusion": conclusion}

# Example usage
sample_mean = 52  # Sample mean
population_mean = 50  # Hypothesized population mean
population_std = 8  # Population standard deviation
sample_size = 36  # Sample size
alpha = 0.05 # significance level

results = z_test(sample_mean, population_mean, population_std, sample_size, alpha)
results
```

Q16. Create a confidence interval for a dataset using Python and interpret the result ?

```python
#Ans.

import numpy as np
import scipy.stats as stats

# Example dataset
data = [22, 24, 23, 20, 25, 27, 29, 26, 28, 30]

# Step 1: Calculate the sample mean and sample standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
n = len(data)

# Step 2: Compute the standard error of the mean
standard_error = sample_std / np.sqrt(n)

# Step 3: Define the confidence level (95% confidence level here)
confidence_level = 0.95
alpha = 1 - confidence_level

# Step 4: Find the critical t-value using the t-distribution (since sample size is small)
t_critical = stats.t.ppf(1 - alpha / 2, df=n - 1)

# Step 5: Calculate the margin of error
margin_of_error = t_critical * standard_error

# Step 6: Compute the confidence interval
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

# Print the results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Standard Error: {standard_error:.2f}")
print(f"Critical t-value: {t_critical:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"95% Confidence Interval: {confidence_interval}")
```

Q17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean ?

```python
#Ans.

# prompt: Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean ?

import numpy as np
from scipy.stats import norm

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculates the confidence interval for the mean of a dataset.

    Args:
        data: A list or NumPy array of numerical data.
        confidence_level: The desired confidence level (default is 0.95).

    Returns:
        A tuple containing the lower and upper bounds of the confidence interval.
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data)
    alpha = 1 - confidence_level
    z_critical = norm.ppf(1 - alpha / 2)  # Find the critical z-value
    margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound


# Example Usage
np.random.seed(42)  # for reproducibility
data = np.random.normal(loc=10, scale=2, size=100)  # Generate sample data from a normal distribution
confidence_level = 0.95

lower_bound, upper_bound = calculate_confidence_interval(data, confidence_level)

print(f"Sample Mean: {np.mean(data)}")
print(f"{confidence_level*100:.0f}% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")


# Interpretation:
# We are {confidence_level*100:.0f}% confident that the true population mean lies within the calculated interval.
```

Q22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution ?

```python
#Ans.

# prompt: Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution ?

import numpy as np

def calculate_range_iqr(data):
    """Calculates the range and interquartile range (IQR) of a dataset.

    Args:
      data: A list or numpy array of numerical data.

    Returns:
      A dictionary containing the range and IQR.
    """
    data = np.array(data)  # Convert to numpy array for easier calculations
    data_range = np.max(data) - np.min(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    iqr = q3 - q1
    return {"range": data_range, "iqr": iqr}

# Example usage with a normal distribution
data = np.random.normal(loc=50, scale=10, size=100) # Example dataset from normal distribution

range_iqr_results = calculate_range_iqr(data)
range_iqr_results
```
